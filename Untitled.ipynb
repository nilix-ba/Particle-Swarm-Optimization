```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

Each particle needs to carry some informations with itself including the most recent position, velocity and the convenience of that position.

```python
class particle:
    
    def __init__(self, position, velocity, cost):
        
        self.position = np.array(position)
        self.best_position = position 
        self.best_cost = cost
        self.v = np.array(velocity)
```

The next position of f which is determined through the formula discussed in class

```python
def f_moves(p):
    
    global w
    global c1
    global c2
    global best_global_f
    
    r1 = random.uniform(0, 2)
    r2 = random.uniform(0, 2)
    
    inertia = p.v.dot(w)
    cognitive = (np.subtract(p.best_position, p.position)).dot(r1*c1)
    total = np.add(inertia, cognitive)
    social = (np.subtract(best_global_f, p.position)).dot(r2*c2)
    total = np.add(total, cognitive)
    
    p.v = total
    p.position = np.add(p.position, p.v)
    
    if (p.position[0] > 10) or (p.position[0] < -10):
        
        p.position[0] =  (p.position[0]%10) - 10
        
    if (p.position[1] > 10) or (p.position[1] < -10):
        
        p.position[1] =  (p.position[1]%10) - 10
```

```python
def g_moves(p):
    
    global w
    global c1
    global c2
    global best_global_g
    
    r1 = random.uniform(0, 2)
    r2 = random.uniform(0, 2)
    
    inertia = p.v.dot(w)
    cognitive = (np.subtract(p.best_position, p.position)).dot(r1*c1)
    total = np.add(inertia, cognitive)
    social = (np.subtract(best_global_g, p.position)).dot(r2*c2)
    total = np.add(total, cognitive)
    
    p.v = total
    p.position = np.add(p.position, p.v)
    
    if (p.position[0] > 100) or (p.position[0] < -100) or (p.position[0] == 0):
        
        p.position[0] = 1 - p.position[0]
        p.position[0] =  (p.position[0]%100) - 100
        
    if (p.position[1] > 100) or (p.position[1] < -100):
        
        p.position[1] = 1 - p.position[1]
        p.position[1] =  (p.position[1]%100) - 100
```

```python
def f(p):
    
    x = p[0]
    y = p[1]
    
    x_2 = x**2
    y_2 = y**2
    
    func_f = abs(1 - ((math.sqrt(x_2 + y_2)) / math.pi))
    func_f = abs(math.sin(x) * math.cos(y) * (math.exp(func_f)))
    
    #print(func_f)
    
    return func_f
```

```python
def g(p):
    
    x = p[0]
    y = p[1]
    
    dev = y/x
    
    func_g = (math.sin(dev)) / (1 + (math.cos(dev))) 
    func_g = (math.pi * math.cos(x) * math.tan(y)) * func_g
    func_g = x * math.sin(func_g)
    
    return func_g
```

```python
def cost_f(p):
    
    global best_global_cost_f
    global best_global_f
    
    func_f = f(p.position)
    
    if (func_f > p.best_cost):

        p.best_cost = func_f   
        p.best_position = p.position
        
    if(func_f > best_global_cost_f):
        
        best_global_cost_f = func_f 
        best_global_f = p.position
        
        print("Best Global position:")
        print(best_global_f)
        print("cost:")
        print(best_global_cost_f)
```

```python
def cost_g(p):
    
    global best_global_cost_g
    global best_global_g
    
    func_g = g(p.position)
    
    if (func_g < p.best_cost):

        p.best_cost = func_g   
        p.best_position = p.position
        
    if(func_g < best_global_cost_g):
        
        best_global_cost_g = func_g
        best_global_g = p.position
        
        print("Best Global position:")
        print(best_global_g)
        print("cost:")
        print(best_global_cost_g)
```

```python
global w
global c1
global c2
global best_global_f
global best_global_cost_f
global gen_f 

gen_f = []

best_global_f = np.array([0,0])
best_global_cost_f = -999

def main_f(n, c1_inp, c2_inp, w_inp):
    
    global w
    global c1
    global c2
    global best_global_f
    global best_global_cost_f
    global gen_f
    
    c1 = c1_inp
    c2 = c2_inp
    w = w_inp
    
    parts = []
    
    for i in range(n):
        
        p_pos = np.array([random.uniform(-10, 10), random.uniform(-10, 10)])
        p_v = np.array([random.uniform(-10, 10), random.uniform(-10, 10)])
        p = particle(position = p_pos, velocity = p_v , cost = f(p_pos))
        
        parts.append(p)
        cost_f(p)
    
    helper = [] #members of a generation   
    for i in range(n):
        
        helper.append(parts[i].position)
    
    gen_f.append(helper[:])
    
    while best_global_cost_f < 19.2:
        
        for part in parts:
            
            f_moves(part)
            cost_f(part)
            
#             print("Best Global position:")
#             print(best_global_f)
#             print("cost:")
#             print(best_global_cost_f)
    
        helper = [] #members a the generation   
        for i in range(n):

            helper.append(parts[i].position)

        gen_f.append(helper[:])
        
        print("Best Global position:")
        print(best_global_f)
        print("cost:")
        print(best_global_cost_f)
```

```python
global w
global c1
global c2
global best_global_g
global best_global_cost_g
global gen_g 

gen_g = []

best_global_g = np.array([0,0])
best_global_cost_g = 999

def main_g(n, c1_inp, c2_inp, w_inp):
    
    global w
    global c1
    global c2
    global best_global_g
    global best_global_cost_g
    global gen_g
    
    c1 = c1_inp
    c2 = c2_inp
    w = w_inp
    
    parts = []
    
    for i in range(n):
        
        p_pos = np.array([random.uniform(-100, 100), random.uniform(-100, 100)])
        p_v = np.array([random.uniform(-100, 100), random.uniform(-100, 100)])
        p = particle(position = p_pos, velocity = p_v , cost = g(p_pos))
        
        parts.append(p)
        cost_g(p)
    
    helper = [] #members of a generation   
    for i in range(n):
        
        helper.append(parts[i].position)
    
    gen_g.append(helper[:])
    
    while best_global_cost_g > (-1.7) * (10**6):
        
        for part in parts:
            
            g_moves(part)
            cost_g(part)
    
        helper = [] #members a the generation   
        for i in range(n):

            helper.append(parts[i].position)

        gen_g.append(helper[:])
        
    print("Best Global position:")
    print(best_global_g)
    print("cost:")
    print(best_global_cost_g)
```
